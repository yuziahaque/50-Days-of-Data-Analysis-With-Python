*2.*

*Using the Matplotlib, plot a bar plot of the sales value of each item. Your plot should have axis labels and a title. The plot size should be 10cm and 8cm.* 

```python
df["Total Sales"] = df["Total Sales"].str.replace("$","").astype(int)
df.dtypes
```

```python
import matplotlib.pyplot as plt
grouped_sales = df.groupby("Item")["Total Sales"].sum()
plt.figure(figsize=(12,8))
plt.bar(grouped_sales.index, grouped_sales)
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.title("Product Sales Analysis", fontsize=20)
plt.show()
```

*3.*

*Using pandas and Matplotlib, plot a pie plot showing the percentage of total sales for each item. Your plot should have a title.* 

```python
grouped_sales = df.groupby("Item")["Total Sales"].sum()

plt.figure(figsize=(10,8))
plt.pie(grouped_sales, labels=grouped_sales.index, autopct='%1.1f%%')
plt.title("Total Sales Analysis",fontsize = 20)
plt.show()
```

*4.*

*Using Matplotlib subplots, plot both a bar plot and a line plot showing total sales for each item. Your subplot must have one column and two rows. The line plot must be on top.* 

```python
grouped_sales = df.groupby("Item")["Total Sales"].sum()

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(15, 10))

ax[1].bar(grouped_sales.index, grouped_sales)

ax[0].plot(grouped_sales.index, grouped_sales)
plt.xlabel("Products")
plt.ylabel("Revenue")
fig.suptitle("Sales Analysis", fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```
