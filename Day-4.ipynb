*4.* 

*Using NumPy, create a 1-dimensional array of 100 random integers from 0 to 10.*

*Use the array to create a histogram of the data and calculate the median and mode.*

*Ensure that the results are reproducible.*

```python
import matplotlib.pyplot as plt
```

```python
randomnum=np.random.default_rng(seed=24)
arr_ran=randomnum.integers(low=0, high=10, size=(100))

arr_ran
```

```python
plt.hist(arr_ran, bins=10,edgecolor='white', color="pink")
plt.title('Histogram')
plt.show()
```

```python
median=np.median(arr_ran)
print('median =', median)
```

```python
from statistics import mode

mode=mode(arr_ran)

print('mode =', mode)
```

*5.*

*Using NumPy, create a 3-dimensional array of 100 random floats between 0 and 1.* 

*Use the array to create a 3D scatter plot of the data.*

*Ensure that your results are reproducible.*

```python
randomnum=np.random.default_rng(seed=24)
arr_ran=randomnum.random(size=(100,3))

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(projection='3d')

ax.scatter(arr_ran[:,0],
           arr_ran[:,1], 
           arr_ran[:,2])

plt.title("3D Scatter Plot", fontsize = 10)
plt.show()
```
