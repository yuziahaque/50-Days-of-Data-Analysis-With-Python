*Import the Text_data dataset. Write a code that checks if there are duplicates in the "text" column.* 

```python
import pandas as pd

df = pd.read_csv("Data/Text_data.csv")
df.head()
```

```python
df.duplicated(subset=['text']).any()
```

*2.*

*Create a copy of the DataFrame and convert the text in the "text" column to lowercase letters.* 

```python
df_copy = df.copy()

df_copy['text'] = df_copy['text'].str.lower()
df_copy.head()
```

*3.*

*Any character that is not a number or in the alphabet is considered to be a special character. Remove special characters in the text column.* 

```python
import string

df_copy['text'] = df_copy['text'].str.translate(str.maketrans('', '', string.punctuation))
df_copy.head()
```

*4.*

*Write a code that adds a column of the length of each row in the text column. Return the row with longest text.* 

```python
df_copy["text length"] = df_copy["text"].str.len()
df_copy.head()
```

```python
df_copy.sort_values(by = ["text length"], ascending=False).iloc[:1]
```

*5.*

*Stopwords are common words that are typically removed from text data before performing natural language processing tasks, such as text classification, sentiment analysis, or information retrieval. Examples of stopwords include "the", "and", "a", "an", "in", "of", "to", etc. Write a code to remove stop words in the text using nltk module* 

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))


df['text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() 
                                                  if word not in stop_words]))
df.head()
```

*6.*

*For text data to be used in machine learning, it must be transformed into numeric vectors. Using scikit-learn, tokenize the text into numeric vectors.* 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

word_count_matrix = cv.fit_transform(df.text)

word_count_df = pd.DataFrame(word_count_matrix.toarray(), columns=cv.get_feature_names_out())
word_count_df.head() 
```
